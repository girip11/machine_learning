## 1. Print output of python statements

By setting `IPython.core.interactiveshell.InteractiveShell.ast_node_interactivity = "all"`, we can print output of all the python statements in a cell. Without this option, we only get the output of the last cell printed.

If we need to do this on all jupyter notebooks, we can add the following lines to the `~/.ipython/profile_default/ipython_config.py` file.

```Python
c = get_config()

c.InteractiveShell.ast_node_interactivity = "all"
```

````python
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"
````

## 2. Display matplotlib output inline

## 3. Line and cell magics

All the line and cell magics supported by ipython can be listed using the line magic `%ls magic`

Useful line magics (start with `%`) and cell magics (start with `%%`)

* `%env` - Lists all environment variables
* `%env MY_ENV=some_value` - Sets the environment variable `MY_ENV`
* `%run <.py|.ipynb file>` - Run an external python file or a jupyter notebook.
* `%load <some_python_script_or_url>` - Replace the contents of the cell with an external python script
* `%who [<data_type>]` - Lists all variables in global scope of the notebook matching the datatype if provided.
* `%timeit` - Times a single python statement using `timeit` python module.
* `%prun <some_python_code>` - Runs the python code and profiles it.
* `%%time` - Runs the code in the cell and times its execution.
* `%%writefile <output_python_file_name>` - Writes the content of that cell to a python file.
* `%pycat <external_python_file>` - Similar to `cat` command in linux. Display contents of python file in that cell with syntax highlighting.
* `!ls` - Execute any shell command by prefixing it with `!`. We could also execute `!pip install some_dependency` from within the notebook.


## 4. Suppress the output of the final statement in a cell

By ending the statement with a semicolon `;`, the output of the last python statement in the cell is suppressed. This trick is often used with plotting to suppress the output from the plot function called.

````python
# No output is displayed after the function gets executed.
np.arange(10);
````

##  6. Run code from different kernel

Using the cell magics like `%%bash`, `%%python3`, `%%ruby` etc, we can run the code written in programming language that is backed by a different jupyter kernel and get the output in the current notebook.

## 9. Using `%pdb` for debugging
 `%pdb` line magic activates python debugger. 
 
 > * When an exception is triggered, IPython can optionally call the interactive pdb debugger after the traceback printout. 
 > * If you want to just activate the debugger AFTER an exception has fired, without having to type `%pdb on` and rerunning your code, you can use the `%debug` magic.

````python
%pdb on


def greet(name: str) -> None:
    print(f"Hello, {name}")
    raise NotImplementedError()


greet("John Doe")
````
