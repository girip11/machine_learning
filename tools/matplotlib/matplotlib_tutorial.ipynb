# Matplotlib tutorial

## Basic plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])

plt.plot(x, y)
plt.xlim(0, 10)
plt.ylim(0, 20)
plt.title("First plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
# size of the figure(canvas) can be modified
# figsize contains the width and the height of the canvas
# width and height specified in inches
plt.figure(figsize=(15, 5), dpi=300)
plt.plot(x, y)

# we can also save the figure
# plt.savefig("mygraph.png", dpi = 300)
plt.show()
```

```python
# customize the line and markers

plt.plot(
    x,
    y,
    color="red",
    marker="^",
    markersize=12,
    markeredgecolor="blue",
    linestyle="dashed",
    linewidth=5,
)
plt.show()
```

```python
# customize the line and markers
# fmt shorthand notation
# [color][marker][line]
plt.plot(x, y, "r^--")
plt.show()
```

```python
# suppose we want to display only the values of x and y on the axes
# we can use ticks to control this
plt.figure(figsize=(15, 5))

plt.plot(
    x,
    y,
    color="red",
    marker="^",
    markersize=12,
    markeredgecolor="blue",
    linestyle="dashed",
    linewidth=5,
)

# Notice the axes have only the values present in the array x and y
# ticks can be any array with required granularity
plt.xticks(x)
plt.yticks(y)

plt.show()
```

```python
# plotting lines with different color and representation
# by default all the points are plotted and line is drawn as solid blue
# This will draw the line as red solid line
plt.plot(x, y, "r-")
plt.show()
```

```python
# points are plotted as green dots
plt.plot(x, y, "go")
plt.show()
```

```python
# we can plot multiple sets of data
x2 = np.arange(1, 5)
y2 = x2 ** 3

plt.figure(figsize=(15, 5))
plt.plot(x, y, "go", x2, y2, "r^")
plt.show()
```

## Multiple plots

```python
# subplot method can be used to add more than one plots in one figure

plt.suptitle("Subplots")
# we will have 1 row and 2 columns and this subplot should occupy
# index 1
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], "go")
plt.title("Squares")
plt.show()

plt.subplot(1, 2, 2)
plt.plot([1, 2, 3, 4], [1, 8, 27, 64], "r^")
plt.title("Cubes")
plt.show()
```

```python
plt.suptitle("Subplots")
# we will have 2 rows and 1 column and this subplot should occupy
# index 1
plt.subplot(2, 1, 1)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], "go")
plt.title("Squares")
plt.show()

plt.subplot(2, 1, 2)
plt.plot([1, 2, 3, 4], [1, 8, 27, 64], "r^")
plt.title("Cubes")
plt.show()
```

```python
# Using subplots method to organize plots better
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))

# you can index the graphs using the ax matrix
ax[0, 0].plot([1, 2, 3, 4], [1, 4, 9, 16], "go")
ax[0, 0].set_title("Squares")

ax[0, 1].plot([1, 2, 3, 4], [1, 8, 27, 64], "b-")
ax[0, 1].set_title("Cubes")

ax[1, 0].plot([1, 2, 3, 4], [2, 4, 6, 8], "r-")
ax[1, 0].set_title("Doubled")

ax[1, 1].plot([1, 2, 3, 4], [3, 6, 9, 12], "y^")
ax[1, 1].set_title("Tripled")

plt.show()
```

## Bar graphs

```python
subjects = np.array(["Maths", "History", "Science", "English", "Geography"])
marks = np.array([100, 60, 90, 80, 70])

print(marks.var())

bars = plt.bar(subjects, marks, color="red")

bars[0].set_hatch("/")
bars[1].set_hatch("*")
bars[2].set_hatch("o")

plt.title("Subjects and marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
# horizontal bar graph
# xerr - horizontal bar graphs with errorbars

xerror = [0, 20, 0, 10, 15]

plt.barh(subjects, marks, xerr=xerror, color="red")
plt.title("Subjects and marks")
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.show()
```

## Pie charts

```python
cloud = np.array(["AWS", "Azure", "GoogleCloud", "Others"])
market_share = np.array([65, 22, 8, 5])

explode = [0.2, 0, 0, 0]

plt.figure(figsize=(10, 5))

plt.pie(market_share, explode=explode, labels=cloud, shadow=True, startangle=45)
plt.axis("equal")

# loc - used to indicate the placement of legend on the figure.
plt.legend(title="Public clouds", loc="best")
plt.show()
```

## Histogram

```python
# randn generates data with the properties of normal distribution
# So the histogram might resemble normal distribution curve
x = np.random.randn(1000)

plt.title("Histogram")
plt.xlabel("Random data")
plt.ylabel("Frequency")

# draws a total to 5 equal width bins
plt.hist(x, 5)
plt.show()
```

## Scatterplot

```python
# scatter plots helpful in identifying linear relationships
x = np.arange(1, 101)
y = np.random.randn(100)

plt.scatter(x, y)
plt.title("Scatterplot")
plt.show()
```

## Plotting example data

```python
import pandas as pd

df = pd.read_csv("WorldHappiness2018_Data.csv")
df.head()
```

```python
rank = df["Rank"]
score = df["Score"]

# plot as line graph
plt.plot(rank, score)
plt.title("World happiness score by rank")
plt.xlabel("Rank")
plt.ylabel("Score")
plt.show()
```

```python
# multiple plots
gdp = df["GDP_Per_Capita"]
life_exp = df["Healthy_Life_Expectancy"]

# We need to provide labels so that legend can display those labels
plt.plot(rank, gdp, color="green", label="Rank vs GDP")
plt.plot(rank, life_exp, color="red", label="Rank vs Life expectancy")
plt.title("Happiness rank vs GDP, life expectancy")
plt.xlabel("Country rank")
plt.legend()
plt.show()
```

```python
# scatter plot
plt.scatter(gdp, score)
plt.ylim(0, 10)
plt.title("GDP vs happiness score")
plt.xlabel("GDP per capita")
plt.ylabel("Happiness score")
plt.show()
```

```python
# histogram
plt.hist(score)
plt.title("Happiness score distribution")
plt.xlabel("Happiness score")
plt.ylabel("Frequency")
plt.show()
```

```python
# bar graphs
rounded_hap_score = score.apply(int)
counts = rounded_hap_score.value_counts()

plt.bar(counts.index, counts)
plt.title("Happiness score spread")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```
