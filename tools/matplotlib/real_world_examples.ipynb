## Gas prices in USD

```python
gas_prices_df = pd.read_csv("gas_prices.csv")
gas_prices_df.head()
```

```python
# Plot gas prices of few countries in the same graph
plt.figure(figsize=(15, 5))
plt.plot(gas_prices_df.Year, gas_prices_df.USA, label="USA gas price")
plt.plot(gas_prices_df["Year"], gas_prices_df.Canada, label="Canada gas price")
plt.plot(gas_prices_df["Year"], gas_prices_df.Japan, label="Japan gas price")
plt.title("Gas price over the years in USD")
plt.xticks(gas_prices_df.Year[::3])
plt.xlabel("Years")
plt.ylabel("Gas Price in $")
plt.legend()
plt.show()
```

```python
# plot all the countries
plt.figure(figsize=(15, 10))

for col in gas_prices_df.columns[1:]:
    plt.plot(gas_prices_df.Year, gas_prices_df[col], label=col, marker=".")

plt.title("Gas price over the years in USD")
plt.xticks(gas_prices_df.Year[::2])
plt.xlabel("Years")
plt.ylabel("Gas Price in $")
plt.legend(loc="best")
plt.show()
```

## Fifa data

```python
fifa = pd.read_csv("fifa_data.csv")
print(fifa.size)
fifa.head()
```

```python
# Histogram of player overall skill level
bins = range(80, 101, 5)
plt.hist(fifa.Overall, bins)
plt.xticks(bins)
plt.yticks(range(0, 500, 50))
plt.xlabel("Skill level")
plt.ylabel("Distribution of player skills")
plt.show()
```

```python
# plot lefties to righties pie chart
leftie = fifa[fifa["Preferred Foot"] == "Left"].count()[0]
rightie = fifa[fifa["Preferred Foot"] == "Right"].count()[0]

plt.title("Fifa players preferred foot distribution")
plt.pie([leftie, rightie], labels=["Left", "Right"], colors=["blue", "green"], autopct="%.2f %%")
plt.legend()
plt.show()
```

```python
# team wise player skill comparison
plt.style.use("default")
barca = fifa[fifa.Club == "FC Barcelona"]["Overall"]
madrid = fifa[fifa.Club == "Real Madrid"]["Overall"]
revs = fifa[fifa.Club == "New England Revolution"]["Overall"]

labels = ["Barcelona", "Real Madrid", "NE Revolution"]

boxes = plt.boxplot([barca, madrid, revs], labels=labels)

for box in boxes["boxes"]:
    box.set(color="blue", linewidth=2)


plt.title("Fifa teams comparison")
plt.ylabel("Club Players overall rating")
plt.show()
```
