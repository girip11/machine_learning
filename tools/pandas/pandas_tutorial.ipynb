# Pandas Tutorial

Pandas has built-in two dimensional data structure called a Dataframe.

## Pandas Series

Series is similar to numpy arrays, except we can give named or datetime as index instead of just a numerical index.

* Numpy arrays should contain only values of the same data type.
* Pandas series are highly flexible

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b', 'c']
values = [ 10, 20, 30]

# pandas series are very similar to numpy arrays
pandas_series = pd.Series(values, index=labels)
pandas_series
```

```python
# we can use labels as index to refer to the value
pandas_series["a"]
```

```python
# we can also use the numerical index
pandas_series[0]
```

```python
# we can also create a pandas series from python dictionary
series_from_dict = pd.Series({"a":10,"b": 20, "c":30})
series_from_dict
```

## Pandas dataframe

* Pandas dataframes are similar to numpy 2dimensional arrays but with much more functionality.
* Pandas dataframe is a two-dimensional data structure that has labels for both its rows and columns.
* Pandas dataframe can be thought of as collection of pandas series

```python
rows = ["X", "Y", "Z"]
columns = ["A", "B", "C", "D", "E"]

values = np.random.default_rng().integers(1,100, (3,5))
print(values)

df = pd.DataFrame(values, rows, columns)
df.head()
```

## Indexing and assignment in Pandas dataframes

```python
# we can refer to any column using the column labels inside []
df["A"]
```

```python
# we can also select a list of columns
df[["A", "B"]]
```

```python
# You can also select a specific element of a specific row 
# using chained square brackets.
df["A"]["X"]
```

## Create and remove columns in a dataframe

```python
# adding a column with label "A+B" from existing columns A and B
df["A+B"] = df["A"] + df["B"]
df["A+B"]
```

```python
# by default drop will drop rows
# drop does not modify the original dataframe
# it returns a dataframe with the specific column removed
df.drop("X")
# deletes row with index X
df.drop(index = ["X"])
```

```python
# To remove a column we should set the axis to 1
df.drop("A+B", axis=1)

# This is easier and more readable. No need to remember the axis
df.drop(columns = ["A+B"])
```

```python
# below code deletes the column from the dataframe inplace
# df.drop("A+B", axis=1, inplace=True)
# or we can overwrite the variable reference to modified dataframe
df = df.drop("A+B", axis=1)
df
```

## Conditional selection using pandas dataframe

```python
# returns a boolean dataframe where each entry is either True or False
# very similar to numpy arrays and conditions
df > 50
```

```python
# we can also use the condition inside []
# when the condition is not satisfied by the entry
# the entry is replaced with NaN
df[(df > 50) & (df < 75)]
```

```python
# select specific column and apply condition
print(df["C"] < 75)

# we can apply condition to any subset of dataframe
df.iloc[:, 1:4] < 75
```

```python
# This will select enties as per the boolean matrix returned by the condition
print(df[df.iloc[:, 1:4] < 75])

# condition applied to subset of dataframe
df.iloc[:, 1:4][df.iloc[:, 1:4] < 75]
```

```python
# rows corresponding to the output of the condition is returned
df[(df['C'] > 50) & (df['A'] < 75)]
```

## Deal with missing data in the pandas dataframes

```python
data = np.array([[1, 5, 1],[2, np.nan, 2],[np.nan, np.nan, 3]])

df = pd.DataFrame(data)
df.columns = ['A', 'B', 'C']

df
```

```python
# dropna method removes any rows that contain NaN value
# Note that like the other DataFrame operations that we have explored,
# dropna does not modify the original DataFrame unless you either 
# (1) force it to using the = assignment operator or 
# (2) specify inplace=True.

df.dropna()
```

```python
# drop only the columns with missing values by setting the axis = 1
df.dropna(axis = 1)
```

```python
# In many cases, you will want to replace missing values 
# in a pandas DataFrame instead of dropping it completely. 
# The fillna method is designed for this.
df.fillna(0) # fills NaN with 0
```

```python
# More often we will replace a missing value with either:
# The average value of the entire DataFrame
# The average value of that row of the DataFrame

df.fillna(df.mean())
```

```python
col_a_mean = df["A"].mean()
print(col_a_mean)
# compute the mean, fill and replace the column
df["A"] = df["A"].fillna(col_a_mean)
df
```

##  `groupby` feature

```python
df = pd.DataFrame([ ['Google', 'Sam', 200],

                    ['Google', 'Charlie', 120],

                    ['Salesforce','Ralph', 125],

                    ['Salesforce','Emily', 250],

                    ['Adobe','Rosalynn', 150],

                    ['Adobe','Chelsea', 500]])

df.columns = ['Organization', 'Salesperson Name', 'Sales']

df
```

```python
df.groupby("Organization").mean("Sales")
```

```python
df.groupby("Organization").sum("Sales")

# we also have count, min, max, std(standard deviation) methods
# documentation on methods available on the groupby object can be
# found at 
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.core.groupby.GroupBy.apply.html
```

```python
# the describe method, which returns useful information 
# for every category that the 
# groupby function is working with.
df.groupby("Organization").describe()
```

## `concat` method

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],

                        'B': ['B0', 'B1', 'B2', 'B3'],

                        'C': ['C0', 'C1', 'C2', 'C3'],

                        'D': ['D0', 'D1', 'D2', 'D3']},

                        index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],

                        'B': ['B4', 'B5', 'B6', 'B7'],

                        'C': ['C4', 'C5', 'C6', 'C7'],

                        'D': ['D4', 'D5', 'D6', 'D7']},

                         index=[4, 5, 6, 7]) 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],

                        'B': ['B8', 'B9', 'B10', 'B11'],

                        'C': ['C8', 'C9', 'C10', 'C11'],

                        'D': ['D8', 'D9', 'D10', 'D11']},

                        index=[8, 9, 10, 11])
```

```python
# takes axis = 0  by default
pd.concat([df1, df2, df3])
```

```python
# along axis = 1
pd.concat([df1, df2, df3], axis = 1)
```

## Merging dataframes(sql inner join)

You can merge two pandas DataFrames along a common column using the merge columns. For anyone that is familiar with the SQL programming language, this is very similar to performing an join in SQL.

```python
leftdf = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K4'],

                     'A': ['A0', 'A1', 'A2', 'A3'],

                     'B': ['B0', 'B1', 'B2', 'B3']})
leftdf
```

```python
rightdf = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],

                          'C': ['C0', 'C1', 'C2', 'C3'],

                          'D': ['D0', 'D1', 'D2', 'D3']})
rightdf
```

```python
pd.merge(leftdf, rightdf, how="inner", on="key")
```

## Pandas `join` method

Commonly used to join dataframes based on index column(column containing row labels). But we can also specify a key column for join.

```python
leftdf = pd.DataFrame({  'A': ['A0', 'A1', 'A2', 'A3'],

                                'B': ['B0', 'B1', 'B2', 'B3']},

                                index =['K1', 'K2', 'K0', 'K4'])
leftdf
```

```python
rightdf = pd.DataFrame({ 'C': ['C0', 'C1', 'C2', 'C3'],

                                'D': ['D0', 'D1', 'D2', 'D3']},

                                index = ['K1', 'K2', 'K0', 'K3']) 
rightdf
```

```python
# join works on index column
# by default on = index column
# Performs left join by default
leftdf.join(rightdf, how="inner")
```

```python
# sort by the join on column
leftdf.join(rightdf, how="inner", sort = True)
```

## Other commons operations in dataframe

```python
df = pd.DataFrame({'col1':['A','B','C','D'],

                   'col2':[2,7,3,7],

                   'col3':['fgh','rty','asd','qwe']})
df
```

```python
# unique values in a pandas series
# unique method works only on series and not on dataframe
df["col2"].unique()
```

```python
# count of unique items
df["col2"].nunique()
```

```python
# count of occurence of each value
df["col2"].value_counts()
```

```python
# apply method can be used to apply a builtin or a custom function 
# on a pandas series
def square(x):
    print(x)
    return x ** 2

df["col2"].apply(square)
```

```python
df["col3"].apply(len)
```

```python
# sort dataframe based on column values using sort_values function
# to modify the dataframe inplace using option inplace = True
df.sort_values("col2")
```
