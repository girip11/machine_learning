## Setting up pytorch using pipenv
* Follow the steps for setting up CUDA on [Cuda installation on ubuntu 20.04](https://varhowto.com/install-pytorch-ubuntu-20-04/)
* Type `nvidia-smi` and find the cuda version from the output. Output would look like this
![cuda version](./cuda_version.png)

* In my case the cuda version is 11. In the [pytorch local installation page](https://pytorch.org/get-started/locally/) select the configuration that would look like ![pytorch local installation](./pytorch_setup.png)

Copy the installation command from the webpage that would look something like below.

```bash
pip install torch==1.7.1+cu110 torchvision==0.8.2+cu110 torchaudio===0.7.2 -f https://download.pytorch.org/whl/torch_stable.html
```

* Visit [pytorch whl files](https://download.pytorch.org/whl/torch_stable.html) and find the whl file url for the **torch** and **torchvision** versions. In my case the versions are `torch==1.7.1+cu110` and `torchvision==0.8.2+cu110`

* Add the following lines to pipenv's Pipfile. Or execute `pipenv install https://download.pytorch.org/whl/cu110/torch-1.7.1%2Bcu110-cp38-cp38-linux_x86_64.whl` and `pipenv install "https://download.pytorch.org/whl/cu110/torchvision-0.8.2%2Bcu110-cp38-cp38-linux_x86_64.whl"`.

```toml
[packages]
torch = {file = "https://download.pytorch.org/whl/cu110/torch-1.7.1%2Bcu110-cp38-cp38-linux_x86_64.whl"}
torchvision = {file = "https://download.pytorch.org/whl/cu110/torchvision-0.8.2%2Bcu110-cp38-cp38-linux_x86_64.whl"}
```

Or you could execute the command `pipenv install fastai ...`

````python
import os

import GPUtil as GPU
import humanize
import psutil

GPUs = GPU.getGPUs()
# XXX: only one GPU on Colab and isn’t guaranteed
gpu = GPUs[0]


def printm():
    process = psutil.Process(os.getpid())
    print(
        "Gen RAM Free: " + humanize.naturalsize(psutil.virtual_memory().available),
        " | Proc size: " + humanize.naturalsize(process.memory_info().rss),
    )
    print(
        "GPU RAM Free: {:.0f}MB | Used: {:.0f}MB | Util {:3.0f}% | Total {:.0f}MB".format(
            gpu.memoryFree, gpu.memoryUsed, gpu.memoryUtil * 100, gpu.memoryTotal
        )
    )


printm()
````
