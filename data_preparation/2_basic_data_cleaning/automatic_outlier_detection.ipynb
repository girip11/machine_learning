# Automatic outlier detection

A one class classifier is used to detect outliers in the dataset.

Examples of such one class classification algorithms are
* LocalOutlierFactor (LOF)
* IsolationForest

In Local Outlier Factor, each training example is assigned a score based on how isolated it is, in otherwords how far it is from its neighbours.

If the training example has higher score, then it is more likely to be an outlier.

## Prediction Without outliers removal

Here we use Linear Regression on the Boston housing dataset.

```python
import pandas as pd
```

```python
df = pd.read_csv("boston_housing.csv", header=None)

# separate features and target variable
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, regressor.predict(X_test))
print(f"Error without outlier removal: {mae}")
```

## Prediction with outliers removed using LOF

Here before creating the model, outliers are removed from the training set. The same parameters learnt on outliers will be applied to the test set for model performance evaluation.

Here only the training data is subjected to outlier detection and removal using LOF algorithm.

```python
df = pd.read_csv("boston_housing.csv", header=None)

# separate features and target variable
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
# Split before outlier detection to prevent data leakage
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
from sklearn.neighbors import LocalOutlierFactor

# LocalOutlierFactor is a predictor
# Remember LOF is a one class classifier
# -1 for outlier, 1 - inlier
lof_outlier_detector = LocalOutlierFactor()
outliers = lof_outlier_detector.fit_predict(X_train)

print(outliers)
```

```python
# Take only the training instances that are not outliers
print(f"Count before outlier removal: {len(X_train)} {len(y_train)}")
X_train, y_train = X_train.loc[outliers == 1], y_train[outliers == 1]
print(f"Count after outlier removal: {len(X_train)} {len(y_train)}")
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Note that the test data set is not subjected to outlier removal
mae = mean_absolute_error(y_test, regressor.predict(X_test))
print(f"Mean absolute error with outlier removal: {mae}")
```
