```python
import numpy as np
import pandas as pd
```

## Identify and delete Single value columns

* Columns with single value are referred to as zero variance predictors.
* Such features useless for modelling as they do not contain any useful information.
* Delete the zero variance columns from the dataset.

```python
df = pd.read_csv("oil_spill.csv", header=None)
print(df.shape)
```

```python
for col in range(len(df.columns)):
    print(f"Column {col} unique values :{len(df.iloc[:, col].unique())}")
```

```python
# remove the zero variance predictors. In other words, remove
# columns containing only one unique value throughout the dataset
unique_counts = df.nunique()

zero_var_cols = []
for index, value in zip(unique_counts.index, unique_counts.values):
    if value == 1:
        zero_var_cols.append(index)

print(f"Columns to delete: {zero_var_cols}")
if zero_var_cols:
    df.drop(zero_var_cols, axis=1, inplace=True)
df.shape
```

## Columns with very few unique values

Find columns that have very few unique values. We can identify those columns by calculating the **ratio of the unique values to the total training samples(rows)**. Those columns with this ratio less than 1% can be identified and those columns can be subjected to one of the following actions

* Delete all those rows
* Can those unique values be encoded as ordinal values?
* Can those unique values be encoded as categorical values?
* Compare model skill with each variable removed from the dataset.

```python
df = pd.read_csv("oil_spill.csv", header=None)
print(df.shape)
```

```python
# Find the unique values percent against the total records
unique_counts = df.nunique()
total_records = len(df.index)
near_zero_variance_cols = []

# Take those columns whose unique value percentage is less than 1%
for index, value in zip(unique_counts.index, unique_counts.values):
    uniq_value_percentage = value / total_records * 100
    if uniq_value_percentage < 1.0:
        near_zero_variance_cols.append(index)
        print(f"{index}, {value}, {uniq_value_percentage:.3f}")
```

```python
# Suppose we decide to delete all the rows
# we can delete using the drop method on the dataframe
print(near_zero_variance_cols)

df.drop(near_zero_variance_cols, axis=1, inplace=True)
df.shape
```

## Remove columns with low variance

* Column with only single unique value will have zero variance. 
* Column with very few unique values compared to the total training samples will have low variance.
* We can configure some variance threshold and remove columns that have variance below the selected threshold.


In scikit learn, `VarianceThreshold` can be used for this purpose.  `VarianceThreshold` by default removes columns with 0.0 variance. But custom threshold can be passed as a parameter.

```python
from sklearn.feature_selection import VarianceThreshold

df = pd.read_csv("oil_spill.csv", header=None)
print(df.shape)

X = df.iloc[:, :-1]
y = df.iloc[:, -1]

variance_thresholds = np.arange(0.0, 0.55, 0.05)
feature_counts = []

for t in variance_thresholds:
    var_threshold = VarianceThreshold(threshold=t)
    X_sel = var_threshold.fit_transform(X)
    # select the column count
    feature_counts.append(X_sel.shape[1])

feature_counts
```

```python
# plot the variance threshold to feature variables
from matplotlib import pyplot as plt

plt.plot(variance_thresholds, feature_counts, "^-")
plt.show()
```
