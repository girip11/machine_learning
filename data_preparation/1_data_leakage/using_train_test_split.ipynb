## Using Naive data preparation

Here we apply the data preprocessing steps to the entire dataset and then split the dataset into training and test set.

Due to this features of test data set gets leaked to the model. This affects the model performance during model evaluation.

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# data preprocessing using the entire dataset
from sklearn.preprocessing import MinMaxScaler
```

```python
def get_dataset():
    return make_classification(
        n_samples=1000, n_features=20, n_informative=15, n_redundant=5, random_state=7
    )
```

```python
# Apply feature scaling  - Normalization on X
# X_min, X_max will include the data from the test set as well
# thus causing a data leakage
def apply_feature_scaling(scaler, features):
    return scaler.transform(features)
```

```python
def create_model(cls, feature_matrix, target_variable):
    model = cls()
    model.fit(feature_matrix, target_variable)
    return model
```

```python
# Naive data preparation example
X, y = get_dataset()
print(X.shape, y.shape)

# apply feature scaling
scaler = MinMaxScaler()
scaler.fit(X)
X = apply_feature_scaling(scaler, X)

# Now split the dataset into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)

# create and train model
model_with_data_leakage = create_model(LogisticRegression, X_train, y_train)

# predict
yhat = model_with_data_leakage.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, yhat)
print(f"Accuracy: {accuracy * 100}")
```

```python
# Without data leakage
X, y = get_dataset()
print(X.shape, y.shape)

# First split the dataset into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)

# apply feature scaling
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train = apply_feature_scaling(scaler, X_train)

# create and train model
model_with_data_leakage = create_model(LogisticRegression, X_train, y_train)

# Before predicting, apply feature scaling to test set
# based on parameters computed on the training set.
X_test = apply_feature_scaling(scaler, X_test)
# predict
yhat = model_with_data_leakage.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, yhat)
print(f"Accuracy: {accuracy * 100}")
```
