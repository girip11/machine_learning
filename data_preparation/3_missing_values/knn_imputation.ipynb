# KNN Imputation

We can also use models to predict missing values in columns. One such model is K-nearest neighbours. Such imputation is known as nearest neighbour imputation.

`KNNImputer` has hyperparameters like K(how many neighbours), what metric should be used for distance calculation.

```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
df = pd.read_csv("horse_colic.csv", header=None, na_values="?")

display(df.head(5))
df.info()
```

```python
X = df.drop([23], axis=1).values
y = df.iloc[:, [23]].values

print(f"{X.shape}, {y.shape}")
```

```python
print(f"Missing values before imputation: {sum(np.isnan(X).flatten())}")
```

```python
from sklearn.impute import KNNImputer

# should not consider NaN values for distance computation
imputer = KNNImputer(n_neighbors=5, metric="nan_euclidean")

X = imputer.fit_transform(X)
```

```python
print(f"Missing values before imputation: {sum(np.isnan(X).flatten())}")
```

Better results can be obtained by tuning/changing the number of neighbours and evaluating the model.
