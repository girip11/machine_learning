# Iterative imputation

In this type of imputation, each feature is considered as a function of all other features and its missing values are predicted like a regression problem.

As we keep moving from one feature to another feature, in one iteration all of the features will be filled some values.

Using this as the input for second iteration, each of the feature's missing values are predicted from all other features but with their missing values being filled/estimated in the previous iteration.

Usually the number of iterations is kept to a small number like 10(this is a hyperparameter). 

Also imputation order is another hyperparameter. Commonly we proceed from filling the features with least missing values to most missing values.

Linear models are often used for estimating the missing values. This is another hyperparameter.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("horse_colic.csv", header=None, na_values="?")
df.head(5)
```

```python
# Helps us to view the columns with missing values in one shot
print(df.shape[0])
df.info()
```

```python
print(f"Missing values by columns: {df.isna().sum()}")
```

```python
X = df.drop([23], axis=1).values
y = df.iloc[:, 23].values
print(f"{X.shape}, {y.shape}")
```

```python
print(f"Total values missing before imputation: {np.sum(np.isnan(X).flatten())}")
```

```python
# this is required since IterativeImputer is experimental
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer
from sklearn.linear_model import BayesianRidge

# some important hyperparameters of IterativeImputer
itr_imputer = IterativeImputer(estimator=BayesianRidge(), imputation_order="ascending", max_iter=10)
X = itr_imputer.fit_transform(X)
```

```python
print(f"Total values missing after imputation: {np.sum(np.isnan(X).flatten())}")
```
