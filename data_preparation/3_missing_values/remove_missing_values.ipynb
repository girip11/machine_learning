## Diabetes dataset

Missing values are marked as zero. With domain knowledge we will be able to identify columns for which zero is not a valid value.

For Each Attribute: (all numeric-valued)

   1. Number of times pregnant
   2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
   3. Diastolic blood pressure (mm Hg)
   4. Triceps skin fold thickness (mm)
   5. 2-Hour serum insulin (mu U/ml)
   6. Body mass index (weight in kg/(height in m)^2)
   7. Diabetes pedigree function
   8. Age (years)
   9. Class variable (0 or 1)


```python
df = pd.read_csv("diabetes.csv", header=None)

print(df.shape)
# since all missing values are marked with 0, all the columns
# seems to be non null.
df.info()
```

```python
df.describe()
```

> Missing values are frequently indicated by out-of-range entries; perhaps a negative number (e.g., -1) in a numeric field that is normally only positive, or a 0 in a numeric field that can never 
normally be 0.

0 is invalid for the following columns
* Plasma glucose concentration
* Diastolic blood pressure
* Triceps skin fold thickness
* 2-Hour serum insulin
* Body mass index

```python
# Count missing values for these columns
missing = (df[[1, 2, 3, 4, 5]] == 0).sum()

# We cannot always remove all the rows with missing values
missing
```

```python
rows_with_missing_values = df.loc[
    (df[1] == 0) | (df[2] == 0) | (df[3] == 0) | (df[4] == 0) | (df[5] == 0)
]

print(len(rows_with_missing_values))
```

We cannot always remove the rows with missing values, because in this example dataset we have 376 rows with missing values out of 768 rows. Removing 376 rows means we are just left with very few training samples.

Thus depending on the number of missing values in each column, we may have to handle those columns differently.


```python
# replace missing values or invalid values with np.NaN
# in numpy nan, NaN and NAN are all same and can be used interchangeably
df[[1, 2, 3, 4, 5]] = df.iloc[:, [1, 2, 3, 4, 5]].replace(0, np.NaN)

# isna and isnull methods are aliases
print(df.isna().sum())
print(df.isnull().sum())
```

If invalid(missing values filled with some invalid values) values are fed as part of the data, machine learning algorithms learn those as features and expect the same in the new data. This will impact the performance of the ML algorithm.

Some machine learning algorithms cannot tolerate any missing values.

Remove missing values using `pandas.DataFrame.dropna` is one approach. But this has the negative side of drastically reducing the training examples. Alternatively we can impute the missing values with some statistics like mean, median etc.

```python
df.dropna(inplace=True)
print(df.shape)
```
