## Imputing missing values with statistics

In this we would like to prepare the data to predict whether the horse with colic required surgery or not. Column 23 presents the target variable.

> These values can be expressed in many ways. Iâ€™ve seen them show up as nothing at all, an empty string , the explicit string `NULL` or `undefined` or `N/A` or `NaN`, and the number `0`, among others.

In the horse colic dataset, missing values in the dataset are marked with "?"

```python
df.isnull().sum()
```

```python
# percentage of missing values in each column
missing_values_percentage = (df.isnull().sum() / df.shape[0]) * 100
missing_values_percentage
```

```python
X = df.drop(columns=[23]).values
y = df.iloc[:, 23].values

print(X.shape, y.shape)
print(f"Missing values before imputation: {sum(np.isnan(X).flatten())}")
```

```python
from sklearn.impute import SimpleImputer

# Note the imputation should always be done on the training dataset
# This means we should have split the dataset prior to this step.
imputer = SimpleImputer(strategy="mean")
X = imputer.fit_transform(X)
```

```python
print(f"Missing values after imputation: {sum(np.isnan(X).flatten())}")
```

Other strategies for `SimpleImputer` are `median`, `most_frequent`(mode), `constant`
