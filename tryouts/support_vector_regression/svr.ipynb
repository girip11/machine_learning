```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Position_Salaries.csv")
df.head()
```

```python
X = df.loc[:, ["Level"]].values
print(X)
```

```python
y = df.loc[:, "Salary"].values
print(y)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler()
# We could also do fit_transform(X)
# let the computations be done on the training data
X_scaler.fit(X)

# replace the computed values
X = X_scaler.transform(X)
print(X)
```

```python
# standard scaler expects a 2d array as input
# so we make the dependent variable column as 2d matrix
# and then pass to the scaler
y_scaler = StandardScaler()
y = y_scaler.fit_transform(y.reshape(-1, 1)).flatten()
print(y)
```

```python
# This is just to demonstrate, given a feature matrix,
# scaler applies standardization by calculating required
# statistics on each column of the matrix.
test_sc = StandardScaler()
test_X = test_sc.fit_transform(df.iloc[:, 1:].values)
print(test_X)
```

## Training SVR

```python
from sklearn.svm import SVR

# SVM has so many kernels that can deal with linear
# as well as non linear data.
# refer to this website
# https://data-flair.training/blogs/svm-kernel-functions/
svr = SVR(kernel="rbf")
svr.fit(X, y)
```

## Predicting

```python
# Any parameter for prediction should be on the
# same scale as that of the trained dataset
X_test = X_scaler.transform(np.array([[6.5]]))
X_test
```

```python
test_pred = svr.predict(X_test)
print(test_pred)
```

```python
predicted_salary = y_scaler.inverse_transform(test_pred.reshape(-1, 1)).ravel()
predicted_salary
```

## Visualizing the results

```python
X = X_scaler.inverse_transform(X)
X
```

```python
y = y_scaler.inverse_transform(y.reshape(-1, 1)).ravel()
y
```

```python
plt.scatter(X[:, 0], y, c="r")  # plot original points

# Plot the SVR curve

plt.plot(X, y_scaler.inverse_transform(svr.predict(X_scaler.transform(X))))

plt.show()
```

## Smoother Curve

```python
plt.scatter(X[:, 0], y, c="r")  # plot original points

# we need lot of points for smoother curve
x = np.arange(X[:, 0].min(), X[:, 0].max() + 0.1, 0.1)
y_pred = svr.predict(X_scaler.transform(x.reshape(-1, 1)))
y_pred = y_scaler.inverse_transform(y_pred.reshape(-1, 1)).ravel()
plt.plot(x, y_pred)
plt.show()
```
