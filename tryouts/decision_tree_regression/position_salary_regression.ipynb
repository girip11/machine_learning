```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Position_Salaries.csv")
df.head()
```

```python
# no imputer and scaling required
X = df.loc[:, ["Level"]].values
y = df.loc[:, "Salary"].values
```

```python
# since the dataset is too small, we won't split the dataset
# to training and testing dataset
```

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

```python
regressor.predict([[6.5]])
```

## Visualization

```python
plt.scatter(X, y, c="r")
x = np.arange(X.min(), X.max(), 0.01)
plt.plot(x, regressor.predict(x.reshape(-1, 1)), "g")
plt.show()
```

```python
regressor.score(X, y)
```
