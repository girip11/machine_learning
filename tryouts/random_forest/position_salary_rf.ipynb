```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Position_Salaries.csv")
df.head()
```

```python
X = df.loc[:, ["Level"]].values
y = df.loc[:, "Salary"].values
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X, y)
```

```python
regressor.score(X, y)
```

```python
regressor.predict([[5.5]])
```

## Visualization

```python
plt.scatter(X, y, c="r")
plt.plot(X, regressor.predict(X), "g^-")
plt.show()
```

```python
plt.scatter(X, y, c="r")
x = np.arange(X.min(), X.max(), 0.01)
plt.plot(x, regressor.predict(x.reshape(-1, 1)), color="green")
plt.show()
```
