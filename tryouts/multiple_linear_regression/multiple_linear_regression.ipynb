```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("50_Startups.csv")
df.head()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# help(ColumnTransformer)
ct = ColumnTransformer(
    transformers=[("encoder", OneHotEncoder(), ["State"])], remainder="passthrough"
)

# Use dataframe when you pass column names to the column transformer
X = ct.fit_transform(df.iloc[:, :-1])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(X_test[:, 4], y_test, c="r")
plt.scatter(X_test[:, 4], y_pred, c="g")
plt.show()
```

```python
# print y_test and y_pred side by side
pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
```

```python
# Coefficient of determination
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```
