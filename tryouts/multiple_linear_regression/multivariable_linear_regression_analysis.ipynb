```python
cal_profits = df.Profit[df["State"] == "California"].values
```

```python
def plot_hist(fig, ax, profits):    
    fig.set_figwidth(15)
    fig.set_figheight(5)
    ax.hist(profits, bins = range(0, 200000, 10000))
    ax.set_xticks(range(0, 200000, 10000))
```

```python
cal_profits = df.Profit[df["State"] == "California"].values
fig, ax = plt.subplots()
plot_hist(fig, ax, cal_profits)
```

```python
profits = df.Profit[df["State"] == "Florida"].values
fig, ax = plt.subplots()
plot_hist(fig, ax, profits)
```

```python
profits = df.Profit[df["State"] == "New York"].values
fig, ax = plt.subplots()
plot_hist(fig, ax, profits)
```
