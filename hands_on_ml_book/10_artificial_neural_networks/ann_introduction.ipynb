## Perceptrons

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron

X, y = load_iris(return_X_y=True, as_frame=True)
X.columns
```

```python
# for perceptrons to work effectively, training instances must be linearly separable
# So lets choose only the petal length and petal width features
X.drop(columns=["sepal length (cm)", "sepal width (cm)"], inplace=True)
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, stratify=y, random_state=42
)
```

```python
perceptron = Perceptron()
perceptron.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

y_hat = perceptron.predict(X_test)
accuracy_score(y_test, y_hat)
confusion_matrix(y_test, y_hat)
```
