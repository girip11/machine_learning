## Immutable tensors

```python
arr = tf.constant(range(1, 11), dtype=tf.float64)
arr.shape
```

```python
arr.dtype
```

```python
arr = tf.reshape(arr, (2, 5))
arr
```

```python
# indexing works similar to numpy array
arr[1, 1]
arr[:, 2:4]
arr[..., 1]
```

```python
# adding new axis
t = arr[..., 1, tf.newaxis]
t.shape
t
```

```python
tf.transpose(arr)
```

```python
tf.matmul(arr, tf.transpose(arr))
```

```python
arr ** 2
```

```python
# Calculating max, mean, sum
# Perform max, sum and mean along the first axis(dimension)
# which is along the rows for a matrix
tf.reduce_max(arr, axis=0)
tf.reduce_sum(arr, axis=0)
tf.reduce_mean(arr, axis=0)
```

```python
# mathematical operations
tf.math.log(arr)
```

## Keras low level APIs

```python
# For the code to work in other keras implementations
# we need to use low level operations defined in keras package
from tensorflow.keras import backend as K

# all low level apis are available in keras.backend
K.square(arr)
K.transpose(arr)
```

## Tensors and Numpy

It is possible to create tensors from numpy array and vice versa.

```python
import numpy as np
```

```python
# convert numpy array to tensor
np_arr = np.arange(1, 11, dtype=np.int64)
tf_arr = tf.constant(np_arr)
tf_arr.dtype  # numpy array type is preserved
tf_arr.shape
```

```python
# applying tensorflow operations on numpy arrays
tf.square(np_arr)
# applying numpy operations on tensors
np.square(tf_arr)
```

Tensor operations work on tensors of exactly same datatype. Even adding a float32 and float64 tensors would raise error.

```python
# to add tensors of different types we need to use tf.cast
tf_float_arr = tf.cast(tf_arr, dtype=tf.float32)
tf_arr.dtype, tf_float_arr.dtype
```

## Variables in tensorflow

```python
# initial_value, dtype, shape, trainable are some of the
# important parameters
# shape should match the initial_value shape
tf_var = tf.Variable([[1, 2, 3], [4, 5, 6]], dtype=tf.float64, shape=(2, 3))
```

```python
tf_var.shape
```

```python
# we cannot assign like this to a tensor
# we have to use the assign method
tf_var[1, 2] = 10
```

```python
# variables are mutable
# inplace modification using assign method
tf_var.assign(2 * tf_var)
tf_var
```

```python
# modify a subset
# replace a particular element
tf_var[1, 2].assign(100)
# replace second column with
tf_var[:, 1].assign([1, 1])
tf_var
```

```python
tf_var.scatter_nd_update(indices=[[0, 0], [0, 2]], updates=[300, 400])
tf_var
```
